```python
###### import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a = None

while True:

    print("\n========== Data Analysis & Visualization Program ==========")
    print("1. Load Dataset")
    print("2. Explore Data")
    print("3. Perform DataFrame Operations")
    print("4. Handle Missing Data")
    print("5. Generate Descriptive Statistics")
    print("6. Data Visualization")
    print("7. Save Visualization")
    print("8. Exit")
    print("===========================================================")

    try:
        choice = int(input("Enter your choice: "))

        match choice:

            # ---------------- Load Dataset ----------------
            case 1:
                file = input("Enter CSV file name: ")
                a = pd.read_csv(file, na_values=["nan", "NAN", "NA", "na"])
                print("\nDataset Loaded Successfully!\n")
                print(a)

            # ---------------- Explore Data ----------------
            case 2:

                if a is None:
                    print("Please load dataset first!")
                    continue

                print("\n1. Display first 5 rows")
                print("2. Display last 5 rows")
                print("3. Display column names")
                print("4. Display data types")
                print("5. Display basic information")

                ch1 = int(input("Enter your choice: "))

                if ch1 == 1:
                    print(a.head())

                elif ch1 == 2:
                    print(a.tail())

                elif ch1 == 3:
                    print(a.columns)

                elif ch1 == 4:
                    print(a.dtypes)

                elif ch1 == 5:
                    a.info()

            # ---------------- DataFrame Operations ----------------
            case 3:

                if a is None:
                    print("Please load dataset first!")
                    continue

                print("\n1. Sort Data")
                print("2. Filter Data")

                ch2 = int(input("Enter your choice: "))

                if ch2 == 1:

                    col = input("Enter column name to sort: ")
                    s = a.sort_values(by=col)
                    print(s)

                elif ch2 == 2:

                    col = input("Enter column name: ")
                    value = float(input("Enter filter value: "))
                    print(a[a[col] > value])

            # ---------------- Missing Data ----------------
            case 4:

                if a is None:
                    print("Please load dataset first!")
                    continue

                print("\n==== Handle Missing Data ====")
                print("1. Display rows with missing values")
                print("2. Fill missing values with mean")
                print("3. Drop rows with missing values")
                print("4. Replace missing values with specific value")

                ch3 = int(input("Enter your choice: "))

                if ch3 == 1:
                    print(a[a.isnull().any(axis=1)])

                elif ch3 == 2:

                    col = input("Enter numeric column name: ")
                    a[col] = a[col].fillna(a[col].mean())
                    print(a)

                elif ch3 == 3:

                    print(a.dropna())

                elif ch3 == 4:

                    value = input("Enter replacement value: ")
                    print(a.fillna(value))

            # ---------------- Statistics ----------------
            case 5:

                if a is None:
                    print("Please load dataset first!")
                    continue

                print(a.describe())

            # ---------------- Visualization ----------------
            case 6:

                if a is None:
                    print("Please load dataset first!")
                    continue

                print("\n1. Bar Plot")
                print("2. Line Plot")
                print("3. Histogram")
                print("4. Scatter Plot")
                print("5. Pie Chart")
                print("6. Stack Plot")

                ch4 = int(input("Enter your choice: "))

                if ch4 == 1:

                    x = input("Enter X column: ")
                    y = input("Enter Y column: ")

                    plt.figure(figsize=(10,4))
                    plt.bar(a[x], a[y], color="green", edgecolor="black")
                    plt.xlabel(x)
                    plt.ylabel(y)
                    plt.title("Bar Chart")
                    plt.show(block=False)

                elif ch4 == 2:

                    x = input("Enter X column: ")
                    y = input("Enter Y column: ")

                    plt.figure(figsize=(10,4))
                    plt.plot(a[x], a[y], marker="o", linestyle="--", color="green")
                    plt.xlabel(x)
                    plt.ylabel(y)
                    plt.title("Line Chart")
                    plt.show(block=False)

                elif ch4 == 3:

                    x = input("Enter numeric column: ")

                    plt.figure(figsize=(8,6))
                    plt.hist(a[x], bins=8, color="skyblue", rwidth=0.8)
                    plt.title("Histogram")
                    plt.xlabel(x)
                    plt.ylabel("Frequency")
                    plt.show(block=False)

                elif ch4 == 4:

                    x = input("Enter X column: ")
                    y = input("Enter Y column: ")

                    plt.figure(figsize=(8,6))
                    plt.scatter(a[x], a[y], color="red", edgecolors="black", s=100)
                    plt.xlabel(x)
                    plt.ylabel(y)
                    plt.title("Scatter Plot")
                    plt.show(block=False)

                elif ch4 == 5:

                    x = input("Enter label column: ")
                    y = input("Enter value column: ")

                    plt.figure(figsize=(7,7))
                    plt.pie(a[y], labels=a[x], autopct="%1.1f%%")
                    plt.title("Pie Chart")
                    plt.show(block=False)

                elif ch4 == 6:

                    x = input("Enter X column: ")
                    y1 = input("Enter first Y column: ")
                    y2 = input("Enter second Y column: ")

                    plt.figure(figsize=(10,5))
                    plt.stackplot(a[x], a[y1], a[y2], labels=[y1, y2])
                    plt.legend(loc="upper left")
                    plt.title("Stack Plot")
                    plt.show(block=False)

            # ---------------- Save Plot ----------------
            case 7:

                file = input("Enter file name: ")

                if not file.endswith(".png"):
                    file += ".png"

                plt.savefig(file)
                print(f"Visualization saved as '{file}' successfully!")

            # ---------------- Exit ----------------
            case 8:
                print("Exiting the program... Goodbye!")
                break

            case _:
                print("Invalid Choice!")

    except Exception as e:
        print("Error:", e)
```
